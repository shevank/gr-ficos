```python
import pandas as pd
import matplotlib.pyplot as plt #from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

filename = '/content/drive/MyDrive/Data Science 2022/athleteEventsNoPersonal.csv'     # csv

df = pd.read_csv(filename)   # csv
df
```

```python
# conseguir número de olímpicos por año
numberUniqueYear = df.groupby(['Year'])['ID'].nunique()
```

```python
# esta es una serie de pandas
numberUniqueYear
```

```python
# Puedes convertir a matriz o lista NumPy
uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
#num olympian tambien se debe pasar a solo valores 
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

# **Establece título, etiquetas y límites**

```python
#MAtlab style

# Establecer xlim y ylim
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020) #Puedo establecer de donde a donde en ambos ejes 
plt.ylim(bottom=0,top=1800) #
```

```python
# Establecer xlabel y ylabel 
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year') #Nombre de los ejes
plt.ylabel('Number of Olympians') #Nombre de los ejes
```

```python
# Establecer título
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
plt.title('Number of Olympians Over the Years') #Establecer el titulo del gráfico
```

```python
# Cambiar tamaño de fuente
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Number of Olympians', fontsize = 16) #Establezo fuente en todos. 
plt.title('Number of Olympians Over the Years', fontsize = 16)
```

```python
#modo orientado a objetos

# Establecer xlim y ylim
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
```

```python
# Establecer xlabel y ylabel
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
axes.set_xlabel('Year')
axes.set_ylabel('Number of Olympians')
```
