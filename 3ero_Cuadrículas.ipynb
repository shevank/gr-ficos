```python
import pandas as pd
import matplotlib.pyplot as plt #from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

filename = '/content/drive/MyDrive/Data Science 2022/athleteEventsNoPersonal.csv'     # csv

df = pd.read_csv(filename)   # csv
df
```

```python
# conseguir número de olímpicos por año
numberUniqueYear = df.groupby(['Year'])['ID'].nunique()
```

```python
# esta es una serie de pandas
numberUniqueYear
```

```python
# Puedes convertir a matriz o lista NumPy
uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
#num olympian tambien se debe pasar a solo valores 
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

```python
# Trazar un gráfico que muestre cómo se vería con los valores predeterminados
plt.plot(uniqueYears, numOlympian, c= 'k')
```

MAtlab style

```python
# Agregar una cuadrícula
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid() #para agregar la cuadricula
```

```python
# solo líneas de cuadrícula horizontales
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(axis = 'y') #para eliminar la línea vertical
```

```python
# solo líneas de cuadrícula verticales
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(axis = 'x') #para eliminar la línea horizontal
```

```python
# cambiar el color de las líneas de la cuadrícula, la transparencia y el estilo de línea
# El hecho de que agregues algo no significa necesariamente que sea lo más atractivo visualmente
# Te animo a afinar las cosas
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(c = 'g', 
         alpha = .9,
         linestyle = '-',
         axis = 'x')
```
